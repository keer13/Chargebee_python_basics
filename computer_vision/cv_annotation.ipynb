```python
import cv2
import matplotlib.pyplot as plt
```

```python
#image flipping
img=cv2.imread("dog.jpg")
flipp=cv2.flip(img,1)
flip3=cv2.flip(img,0)

flip2=cv2.flip(img,-1)
cv2.rectangle(img, (50,50), (200,200), (0,255,0), 2)

plt.figure(figsize=(8,8))


plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("original image")
plt.axis('off')


plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(flipp,cv2.COLOR_BGR2RGB))
plt.title("horizontal filpped Image")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(cv2.cvtColor(flip2,cv2.COLOR_BGR2RGB))
plt.title("both filpped Image")
plt.axis('off')


plt.subplot(2,2,4)
plt.imshow(cv2.cvtColor(flip3,cv2.COLOR_BGR2RGB))
plt.title(" Vertical filpped Image")
plt.axis('off')

plt.show()
```
